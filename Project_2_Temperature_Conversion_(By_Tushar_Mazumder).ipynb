```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

celsius = np.linspace(-10, 40, 30)

fahrenheit = (celsius * 9/5) + 32

data = pd.DataFrame({
    "Celsius (°C)": celsius,
    "Fahrenheit (°F)": fahrenheit
})

print("A glimpse of our dataset:\n")
print(data.head())
plt.plot(data["Celsius (°C)"], data["Fahrenheit (°F)"],
         marker="o", color="green")

plt.xlabel("Celsius (°C)")
plt.ylabel("Fahrenheit (°F)")
plt.title("Celsius vs Fahrenheit")
plt.show()

X = data[["Celsius (°C)"]]

y = data["Fahrenheit (°F)"]

model = LinearRegression()
model.fit(X, y)

plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, model.predict(X), color="red", label="Regression line")

plt.xlabel("Celsius (°C)")
plt.ylabel("Fahrenheit (°F)")
plt.title("Regression learns the conversion rule")
plt.legend()
plt.show()

predicted = model.predict(pd.DataFrame({"Celsius (°C)": [25]}))
print(f"The model predicts 25°C = {predicted[0]:.2f} °F")

print("\nModel learned this formula:")
print(f"Fahrenheit = {model.coef_[0]:.2f} × Celsius + {model.intercept_:.2f}")
```
